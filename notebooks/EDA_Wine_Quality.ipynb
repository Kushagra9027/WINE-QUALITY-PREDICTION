### Step 6
Visualize the distribution of wine quality scores.

```python
sns.countplot(x='quality', data=df)
```

### Step 7
Create a correlation heatmap to identify feature relationships.

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
```

### Step 8
Analyze alcohol content across quality levels.

```python
sns.boxplot(x='quality', y='alcohol', data=df)
```

### Step 9
Examine volatile acidity versus wine quality.

```python
sns.boxplot(x='quality', y='volatile acidity', data=df)
```

### Step 10
Investigate sulphates versus wine quality.

```python
sns.boxplot(x='quality', y='sulphates', data=df)
```

### Step 11
Additional sulphates analysis and visualization.

```python
sns.boxplot(x='quality', y='sulphates', data=df)
```

### Step 12
Explore density variation across quality levels.

```python
sns.boxplot(x='quality', y='density', data=df)
```

### Step 13
Analyze pH distribution across quality categories.

```python
sns.boxplot(x='quality', y='pH', data=df)
```

### Step 14
Create pairwise visualizations to identify patterns among key features.

```python
sns.pairplot(
    df[['alcohol','volatile acidity','sulphates','density','quality']],
    hue='quality'
)
```

### Step 15
Group wines into quality categories and compare feature behavior.

```python
df['quality_label'] = df['quality'].apply(
    lambda x: 'Good' if x >= 7 else 'Average' if x >= 5 else 'Poor'
)
sns.boxplot(x='quality_label', y='alcohol', data=df)
```
